DESCRIÇÃO:

BUSCA ONLINE DA COTAÇÃO DE MOEDAS E IBOV

'1) Requisição de api no site:https://docs.awesomeapi.com.br/api-de-moedas para acesso a cotação das moedas utilizadas neste exemplo (USD/EUR/BTC).

'2) Requisição de api no site: https://finance.yahoo.com/YahooFinance para acesso a cotação atual dos indices e ações utilizadas neste exemplo (IBOV e PETR4).

'3) Comandos para request atraves de um botão


'4) Legenda
 
* key      =>    Label
bid              Compra
ask              Venda
varBid           Variação
pctChange        Porcentagem de Variação
high             Máximo
low              Mínimo


```python
import requests
from tkinter import *

def pegar_cotacoes():
    requisicao = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")
    
    cotacao_bovespa = web.DataReader ("^BVSP", data_source="yahoo", start ="01-01-2000", end="12-31-2099")

    requisicao_dic = requisicao.json()

    compra_dolar = requisicao_dic['USDBRL']['bid']
    venda_dolar = requisicao_dic['USDBRL']['ask']
    min_dolar = requisicao_dic['USDBRL']['low']
    max_dolar = requisicao_dic['USDBRL']['high']
    
    compra_euro = requisicao_dic['EURBRL']['bid']
    venda_euro = requisicao_dic['EURBRL']['ask']
    min_euro = requisicao_dic['EURBRL']['low']
    max_euro = requisicao_dic['EURBRL']['high']
    
    compra_btc = requisicao_dic['BTCBRL']['bid']
    venda_btc = requisicao_dic['BTCBRL']['ask']
    min_btc = requisicao_dic['BTCBRL']['low']
    max_btc = requisicao_dic['BTCBRL']['high']

    var_dolar = requisicao_dic['USDBRL']['pctChange']
    var_euro = requisicao_dic['EURBRL']['pctChange']
    var_btc = requisicao_dic['BTCBRL']['pctChange']
    
    
    texto_resposta['text'] = f'''
     
    
'⚡ DOLAR - USD $'        '⚡ EURO - EUR €'         '⚡BITICOIN - BTC ฿'
    
Compra: {compra_dolar}     Compra: {compra_euro}         Compra: {compra_btc}
Venda:{min_dolar}          Venda: {compra_euro}          Venda: {venda_btc}
Min:{min_dolar}            Min:{min_euro}                Min: {min_btc}
Máx:{max_dolar}            Máx:{max_euro}                Máx: {max_btc}
Variação%:{var_dolar}      Variação%: {var_euro}         Variação%: {var_btc}


'⚡ IBOVESPA - Pts'          

Abertura: -                
Min: -                      
Máx: -                    
Variação%: -                   


Aviso de Isenção de Responsabilidade
*Cotações com atraso de até 10 minutos*  

    '''  
    
janela = Tk()
janela.title("MERCADO AGORA")
texto = Label(janela, text="Mercado Financeiro AO VIVO")
texto.grid(column=0, row=0, padx=10, pady=10)

botao = Button(janela, text="ACESSAR", command=pegar_cotacoes)
botao.grid(column=0, row=1, padx=10, pady=10)

texto_resposta = Label(janela, text="")
texto_resposta.grid(column=0, row=2, padx=10, pady=10)

janela.mainloop()

```
